```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.style.use('seaborn-white')
```

```python
def f(x,y):
  val = np.sin(x)**10 + np.cos(10 + y *x) * np.cos(x)
  return val

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
df = pd.DataFrame(X)
Z= f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='blue');
```
