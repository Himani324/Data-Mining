```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), 'o')
plt.plot(x, np.cos(x), '+')
plt.savefig('my_fig_sincos.png')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
plt.subplot(2,1,1)
plt.plot(x, np.sin(x), 'g')
plt.subplot(2,1,2)
plt.plot(x, np.cos(x), '--r')
plt.show()
```
