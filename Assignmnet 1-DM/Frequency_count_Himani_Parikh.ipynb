```python
# Read given file and convert whole text into string.
file = open("1953-Eisenhower.txt", "r")
text = file.read().replace("\n", " ")
file.close()
```

```python
# calculate frequency and prepare dictionary
def f_calcTermFrequency(key):     
    if key in dictionary:
        dictionary[key] = dictionary[key] + 1
    else:
        dictionary.update({key: 1})  
```

```python
# remove unwanted characters like numbers, special charcters.. for frequency_count
special_characters = ["1","2","3","4","5","6","7","8",",",";","'s","&","?", 
                      "(",")","!",".","-","_",":",'"',"'","--"]
for character in special_characters:
		text = text.lower().replace(character,"")

dictionary = {}

# convert sentence into words
words = text.split()
for word in words:
    f_calcTermFrequency(word)

#print the word_index, vocabulary, and frequency in tabular form.
print("{:<10}  {:<18}  {:<10} ".format("w_idx", "vocab", "tfreq"))
i=0;
for key, value in dictionary.items():
    print ("{:<10}  {:<18}  {:<10} ".format(i, key, value))
    i+=1
```
